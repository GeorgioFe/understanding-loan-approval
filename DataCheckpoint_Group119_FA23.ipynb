### Looking at the Missingness Matrix

```python
msno.matrix(df)
```

After looking at the missingness matrix, it seems like the missingness of 'Dependents' and 'Married' could be connected with one another (there are lines that span both columns for multiple rows).
```python
msno.heatmap(df)
```

From the heatmap (and our matrix interpretation above), it seems that the missingness between the Dependents and Married might be related (the value is 0.4 in the heatmap). The other variables don't seem to have relationships (with regards to missingness)

### Handling Missingness

Since there are very few values missing for the Married column, we will just drop those. For the Self Employed and Loan Amount Term categories, there is a dominating value (the mode makes a high percentage of the values in the column), so we will just impute the mode fot these two columns. For the Loan Amount column, we will just impute the mean, as this column is numeric and doesn't seem to have a dominating value. For credit history, we will just leave these as np.nan values, and see how our model does (we cannot impute the mean as it a binary variable, and we don't want to impute the mode as this variable might be a strong indicator of Loan Approval, we may consider deleting these rows later on). It seems like there is a relationship between the missingness of 'Married' and 'Dependants' columns (MAR). We don't know if we should drop these values or use Multiple Imputation (beyond the scope of this course). We will leave these as nan's for now.

```python
# Dropping rows where the Married values are empty
df = df.dropna(subset=['Married'])

# Imputing column values with the mean/mode where necessary
df['Self Employed'] = df['Self Employed'].fillna(df['Self Employed'].mode()[0])
df['Loan Amount Term'] = df['Loan Amount Term'].fillna(df['Loan Amount Term'].mode()[0])
df['Loan Amount'] = df['Loan Amount'].fillna(df['Loan Amount'].mean())
```

We have dealt with the missingness of multiple columns, and will handle the other missing values when we are developing our model!
